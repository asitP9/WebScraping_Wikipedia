```python
import requests
```

### Scraping the below link

```python
link="https://en.wikipedia.org/wiki/List_of_Indian_people_by_net_worth"
```

### converting the content of the website to text

```python
link_content=requests.get(link).text
```

```python
print(link_content)
```

### We wll be using Beautiful Soup and it has a capability to iterate through a HTML code. Basically it converts the html to the xml /beautiful Soup format

```python
from bs4 import BeautifulSoup
```

### lxml is the easier and faster way to handle the HTML code

```python
soup=BeautifulSoup(link_content, 'lxml')
```

```python
soup
```

```python
print(soup.prettify())
```

```python
print("Title of the webpage \n", soup.title)
```

```python
print("Title of the webpage without tag \n", soup.title.string)
```

```python
print("find all 'a' of the webpage \n", soup.find_all("a"))
```

```python
print("find all 'table' of the webpage \n", soup.find_all("table"))
```

```python
print("find all 'table' with particular class name \n")
table_data=soup.find('table', class_='wikitable sortable')
print(table_data)
```

### find all the a tag in the table 

```python
table_link=table_data.find_all("a")
```

```python
table_link
```

### remove the above links from the table

```python
all_as=[]
for links in table_link:
    all_as.append(links.get('title'))  
```
